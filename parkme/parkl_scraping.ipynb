```python
def get_page(url):
    page = requests.get(url)
    if page.status_code == requests.codes.ok:
        return BeautifulSoup(page.content, 'lxml')
```

```python
def data_points (url):
    data = dict ()
    soup=get_page(url)
    content=soup.find_all(attrs = {"class":"lot-info-section"})[0]
    data ['nev'] = content.find_next("h1").text
    addr=soup.find(attrs = {"class":"module-header-address"})
    street=addr.find_next(attrs = {"class":"module-text"})
    data ['utca'] = street.text
    area=street.find_next(attrs = {"class":"module-text"})
    data ['terulet'] = area.text
    tel=soup.find(attrs = {"class":"module-header-phones"}).find(attrs = {"class":"left module-text"})
    data ['telefon'] = tel.text
    price_0=soup.find_all(attrs = {"class":"module-table-row module-no-border"})[0]
    price=price_0.find(attrs = {"class":"right"})
    time=price_0.find(attrs = {"class":"left lot-rate-type"})
    data ['ar1'] = price.text
    data ['ido1'] = time.text.strip()
    price_1=soup.find_all(attrs = {"class":"module-table-row module-no-border"})[1]
    price_2=price_1.find(attrs = {"class":"right"})
    time_2=price_1.find(attrs = {"class":"left lot-rate-type"})
    data ['ar2'] =price_2.text
    data ['ido2'] =time_2.text.strip()
    data ['url'] = url

    return (data)
example=data_points(url3)
```
